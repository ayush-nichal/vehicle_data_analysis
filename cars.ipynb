```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
```

```python
head = df.head()
print(head)
```

```python
info = df.info()
print(info)
```

```python
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["year"], y=df["mpg"], marker="o")
plt.title("MPG Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Miles Per Gallon (MPG)")
plt.show()
```

**The trend line suggests that MPG has generally improved over the years. This aligns with advancements in fuel efficiency, regulations, and automotive technology.**

```python
df["cubicinches"] = pd.to_numeric(df["cubicinches"], errors="coerce")
df["weightlbs"] = pd.to_numeric(df["weightlbs"], errors="coerce")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["cubicinches"], y=df["time-to-60"], alpha=0.7)
plt.title("Effect of Engine Size on 0-60 Time")
plt.xlabel("Engine Size (Cubic Inches)")
plt.ylabel("Time to 60 mph (seconds)")
plt.show()
```

**Larger engine sizes (higher cubic inches) tend to have lower 0-60 times, meaning they accelerate faster. However, there may be some outliers where heavy vehicles with large engines take longer.**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["brand"], y=df["hp"])
plt.title("Horsepower Preference by Country")
plt.xlabel("Country")
plt.ylabel("Horsepower")
plt.show()
```

**US-manufactured vehicles typically have higher horsepower, while European and other regions show a preference for lower-horsepower vehicles. This may be due to differing regulations, fuel efficiency standards, and road infrastructure.**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["hp"], y=df["mpg"], alpha=0.7)
plt.title("Horsepower vs. MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.show()
```

**Higher horsepower generally results in lower MPG, indicating that powerful engines consume more fuel. This inverse relationship is expected in combustion engines.**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["cubicinches"], y=df["mpg"], alpha=0.7)
plt.title("Cubicinches vs. MPG")
plt.xlabel("Cubicinches")
plt.ylabel("MPG")
plt.show()
```

**Vehicles with larger engines (higher cubic inches) tend to have lower MPG, reinforcing the idea that bigger engines consume more fuel.**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["cylinders"], y=df["mpg"])
plt.title("Cylinders vs. MPG")
plt.xlabel("Number of Cylinders")
plt.ylabel("MPG")
plt.show()
```

**Vehicles with more cylinders (e.g., V8 engines) tend to have lower MPG, while those with fewer cylinders (4-cylinder engines) are more fuel-efficient.**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["cubicinches"], y=df["hp"], hue=df["cylinders"], palette="coolwarm")
plt.title("Engine Size vs. Horsepower by Cylinders")
plt.xlabel("Engine Size (Cubic Inches)")
plt.ylabel("Horsepower")
plt.show()
```

**More cylinders typically mean higher horsepower, with V8 engines generally having more power than 4- or 6-cylinder engines.**
